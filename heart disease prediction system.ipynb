```python

import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("heart.csv")
print("First 5 Rows of Dataset")
print(df.head())
print("\nMissing Values")
print(df.isnull().sum())
df = df.dropna()
df = df.drop_duplicates()

print("\nDataset Cleaned Successfully")
print("\nDataset Information")
print(df.info())
print("\nStatistical Summary")
print(df.describe())
heart_counts = df['target'].value_counts()
print("\nHeart Disease Count")
print(heart_counts)

plt.figure(figsize=(6,5))
heart_counts.plot(kind='bar')
plt.title("Heart Disease Analysis")
plt.xlabel("Target")
plt.ylabel("Count")
plt.xticks([0,1], ['Disease', 'No Disease'])
plt.show()


plt.figure(figsize=(6,6))
heart_counts.plot(
    kind='pie',
    autopct='%1.1f%%',
    labels=['Disease', 'No Disease']
)
plt.title("Heart Disease Distribution")
plt.ylabel("")
plt.show()
avg_age = df['age'].mean()
avg_chol = df['chol'].mean()
print("\nFINAL ANALYSIS")
print("Average Age:", round(avg_age,2))
print("Average Cholesterol Level:", round(avg_chol,2))
if heart_counts[1] > heart_counts[0]:
    print("More patients have heart disease.")
else:
    print("More patients do not have heart disease.")

print("Dataset cleaned and analyzed successfully.")
```
